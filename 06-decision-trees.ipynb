# Decision Trees in Python

The iris dataset contains measurements 150 samples of iris flowers (https://en.wikipedia.org/wiki/Iris_(plant)). 
Load the dataset from `iris.csv` as a pandas data frame.
Train a decision tree on the dataset, and use 2/3 of the data for training and 1/3 for testing. What accuracy can you achieve?


```python
import numpy as np
from sklearn import tree
from sklearn import model_selection
import pandas as pd

dat = pd.read_csv("iris.csv")

dat["Name"] = np.array([str(xi) for xi in dat["Name"]])


train,test = model_selection.train_test_split(dat)

xtrain = np.array(train.loc[:,["SepalLength","SepalWidth","PetalLength","PetalWidth"]])
xtest =  np.array(test.loc[:,["SepalLength","SepalWidth","PetalLength","PetalWidth"]])

ytrain =  np.array(train.loc[:,["Name"]])
ytest =  np.array(test.loc[:,["Name"]])

clf = tree.DecisionTreeClassifier()

clf.fit(xtrain,ytrain)

pred = clf.predict(xtest)

#fix some shape problem
xp = ytest.reshape(38) 

np.mean(pred==xp)
```
